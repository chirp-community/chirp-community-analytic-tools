```python
from utils import get_pickle

file_pickle = "/home/iksadnorth/IdeaProjects/analyze_data/pickles/app.pickle"
properties = get_pickle(file_pickle)
```

```python
from utils import classify

table = classify(properties)
properties_Filter, properties_Repository = table['filter'], table['repository']
```

```python
from collections import defaultdict

# 각 repo 쿼리의 평균 실행시간 계산.
table_repository = defaultdict(lambda : {
    'sum_execution-time-milli-seconds' : 0,
    'num' : 0,
})
for item in properties_Repository:
    key = item['method-logging']
    table_repository[key]['sum_execution-time-milli-seconds'] += item['execution-time-milli-seconds']
    table_repository[key]['num'] += 1
```

```python
from utils import draw

x, y = [], []
for k,v in table_repository.items():
    x.append(k)
    y.append(v['sum_execution-time-milli-seconds'] / v['num'])

draw(x,y)
```
